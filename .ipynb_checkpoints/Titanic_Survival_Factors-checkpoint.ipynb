# What Variables Indicate Survival for Titanic Passengers?

## About the Data

```
VARIABLE DESCRIPTIONS:
survival        Survival
                (0 = No; 1 = Yes)
pclass          Passenger Class
                (1 = 1st; 2 = 2nd; 3 = 3rd)
name            Name
sex             Sex
age             Age
sibsp           Number of Siblings/Spouses Aboard
parch           Number of Parents/Children Aboard
ticket          Ticket Number
fare            Passenger Fare
cabin           Cabin
embarked        Port of Embarkation
                (C = Cherbourg; Q = Queenstown; S = Southampton)

SPECIAL NOTES:
Pclass is a proxy for socio-economic status (SES)
 1st ~ Upper; 2nd ~ Middle; 3rd ~ Lower

Age is in Years; Fractional if Age less than One (1)
 If the Age is Estimated, it is in the form xx.5

With respect to the family relation variables (i.e. sibsp and parch)
some relations were ignored.  The following are the definitions used
for sibsp and parch.

Sibling:  Brother, Sister, Stepbrother, or Stepsister of Passenger Aboard Titanic
Spouse:   Husband or Wife of Passenger Aboard Titanic (Mistresses and Fiances Ignored)
Parent:   Mother or Father of Passenger Aboard Titanic
Child:    Son, Daughter, Stepson, or Stepdaughter of Passenger Aboard Titanic

Other family relatives excluded from this study include cousins,
nephews/nieces, aunts/uncles, and in-laws.  Some children travelled
only with a nanny, therefore parch=0 for them.  As well, some
travelled with very close friends or neighbors in a village, however,
the definitions do not support such relations.
```

### What is the overall survival rate (as a percentage of total passengers in our list)? How does the survival rate differ between genders? 

````python
import pandas as pd

PassData = pd.read_csv("titanic.csv")
# PassData.describe()
# print PassData.head()

# Calculate total number of passengers, survivors and dead 
num_passengers = len(PassData)
dead = PassData[PassData.Survived == 0].count()
num_survived = PassData[PassData.Survived == 1].count()[0]

# Calculate total number of survivors and overall survival rate
from __future__ import division

survival_rate = num_survived / num_passengers
print 'Of ' + str(num_passengers) + ' passengers, ' + str(num_survived) + ' survived.'
print 'The survival rate is ' + str(survival_rate) + ' or '+ str("{0:.0f}%".format(survival_rate * 100))

# Count number of males and females and calculate respective survival rates
num_males = PassData[PassData.Sex == 'male'].count()[0]
print 'There were ' + str(num_males) + ' male passengers onboard.'

num_msurvived = PassData[(PassData.Sex == 'male') & (PassData.Survived == 1)].count()[0]
male_survival_rate = num_msurvived / num_males
print 'The survival rate of male passengers was ' + str("{0:.2f}%".format(male_survival_rate * 100))

num_females = PassData[PassData.Sex == 'female'].count()[0]
print 'There were ' + str(num_females) + ' female passengers onboard.'

num_msurvived = PassData[(PassData.Sex == 'female') & (PassData.Survived == 1)].count()[0]
female_survival_rate = num_msurvived / num_females
print 'The survival rate of female passengers was ' + str("{0:.2f}%".format(female_survival_rate * 100))
````

### What is the average age of all passengers onboard? How does survival differ by passenger age? 

````python
# Survivors are younger than the average age, and those who died are older.

avg_age = PassData.Age[PassData.Age.notnull() == True].mean()
print 'The average age of all passengers onboard was ' + str("{0:.2f}".format(avg_age))

avg_age_survived = PassData.groupby('Survived')['Age'].mean()[1]
print 'The average of age of passengers who survived was ' + str("{0:.2f}".format(avg_age_survived))

avg_age_not_survive = PassData.groupby('Survived')['Age'].mean()[0]
print 'The average of age of passengers who did not survive was ' + str("{0:.2f}".format(avg_age_not_survive))
````

### How many passengers are in each of the three classes?

### What is the survival rate for passengers in each of the three classes of service?

````python
class_surv = PassData.groupby('Pclass')['Survived'].mean()

class_surv 
````

````python
# Plot survival rates, with overall survival rate indicated with solid horizontal line

rates = [male_survival_rate, female_survival_rate, class_surv.iloc[0], class_surv.iloc[1], class_surv.iloc[2]]
factors = ['Male', 'Female', '1st_Class', '2nd_Class', '3rd_Class']

surv_data = {'Survival Rate': rates, 'Factor': factors}

surv_factors = pd.DataFrame(data=surv_data)
````

````python
# create linear model to predict survival 

PassData.describe()
````

````python
cols = PassData.columns
     
%R -i PassData,cols colnames(PassData) <- unlist(cols)

%R surv_factors_lm <- data.frame(PassData)

%R colnames(surv_factors_lm)[3] <- "Class"

%R surv_factors_lm$Class <- as.factor(surv_factors_lm$Class)

%R levels(surv_factors_lm$Class) <- c("1st", "2nd", "3rd")
 
%R head(surv_factors_lm)
````

````python
%%R

# Tried a basic linear model, and the residuals vs. fitted values looked strange: two parallel lines, so we tried another method
# Since we're modeling a binary outcome, live or die, using mainly categorical predictors, logistic regression is appropriate
# Distribution of deviance residuals looks ok
# Coefficients give the change in the log-odds per each unit of increase
# Though the model reduces residual deviance compared to a null model that only includes an intercept term, deviance is still high
# Pseudo R^2 test necessary? 
fit <- glm(Survived ~ Class + Sex + Age, data=surv_factors_lm, binomial)
summary(fit, corr=T)
````

````python
%%R

require(MASS)
exp(cbind(coef(fit), confint(fit)))
# Odds of surviving if in 2nd class are 73% lower, holding other vars constant
# Odds of surviving if in 3rd class are 92% lower
# Being male also 92% lower odds
# For each year older, you have about 4% lower odds of survival
````

````python
%%R

# the drop1 function examines model variance changes by dropping variables one at a time. 
manova.survival <- drop1(fit, test="Chisq")

manova.survival


# A logistic regression is said to provide a better fit to the data if it demonstrates an improvement over a model with fewer predictors. This is performed using the likelihood ratio test, which compares the likelihood of the data under the full model against the likelihood of the data under a model with fewer predictors. Removing predictor variables from a model will almost always make the model fit less well (i.e. a model will have a lower log likelihood), but it is necessary to test whether the observed difference in model fit is statistically significant. Given that H0 holds that the reduced model is true, a p-value for the overall model fit statistic that is less than 0.05 would compel us to reject the null hypothesis. It would provide evidence against the reduced model in favor of the current model. 
````

````python
%%R

# Hosmer-Lemeshow goodness of fit test also shows no significant difference between values fitted by our model and the
# observed values.
library(ResourceSelection)

test_factors <- surv_factors_lm[!is.na(surv_factors_lm["Age"]),c(2:3,5:6)]

# test_factors

hoslem.test(test_factors$Survived, fitted(fit))
````

````python
%%R

mosaic_plot <- surv_factors_lm[,c(2:3,5:6)]

mosaic_plot$Survived <- as.factor(mosaic_plot$Survived)

levels(mosaic_plot$Survived) <- c("No", "Yes")

head(mosaic_plot)
````

````python
%%R

library(vcd)
library(grid)

png("./Plots/mosaic.png", width=800, height=800, res=120)
mosaic( ~ Survived + Sex + Class, data=mosaic_plot, highlighting="Sex", highlighting_fill=c("pink", "lightblue"), 
       direction=c("v", "h", "v"))
dev.off()

mosaic( ~ Survived + Sex + Class, data=mosaic_plot, highlighting="Sex", highlighting_fill=c("pink", "lightblue"), 
       direction=c("v", "h", "v"))
# Can see that a huge portion of those who died were males traveling in 3rd class, and that males dominated the other
# classes of folks who died. Interestingly, more 3rd class males survived than 1st or 2nd, though as there were many 
# more 3rd class passengers, the 3rd class survival rate is much lower. Survivors were mostly female, slightly more 
# first class than others, though not by a large margin.  
````

````python
%%R 

library(survival)

surv_curv <- mosaic_plot

surv_curv$SurvObj <- with(surv_curv, Surv(Age, Survived == "No"))

km.as.one <- survfit(SurvObj ~ 1, data = surv_curv, conf.type = "log-log")
km.by.sex <- survfit(SurvObj ~ Sex, data = surv_curv, conf.type = "log-log")
km.class <- survfit(SurvObj ~ Class, data = surv_curv, conf.type= "log-log")
km.class.sex <- survfit(SurvObj ~ Class + Sex, data = surv_curv, conf.type= "log-log")
````

````python
%%R
library(survminer)

png("./Plots/Surv_Prob_All.png", width=800, height=800, res=120)
print(ggsurvplot(km.as.one, xlab="Age", ggtheme=theme_bw()))
dev.off()

ggsurvplot(km.as.one, xlab="Age", ggtheme=theme_bw())
````
